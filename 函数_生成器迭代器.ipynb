- 生成器必然是一个迭代器，那么什么是生成器呢
- 简单的说，一边循环一边计算生成数的机制，称为生成器：generator，目的是为了防止一次性生成过大的空间，造成浪费或者内存不足。
- 有两种方法产生生成器
   - 1 生成器表达式
   - 2 生成器函数

```python
# 生成器表达式,注意用圆括号括住
it = (item**2 for item in range(9))
type(it)
```

```python
next(it)#注意，遍历生成器，用for循环，不会出现迭代到结尾stop Iteration的报错
```

```python
#需要注意生成器表达式 和列表推导式的区别。用方括号[]代表列表，
#列表推导式：从一个列表中生成一个新的列表，简化了循环的写法
l1= [x for x in range(30) if x%3==0]
l1

#字典推导式：从能生成k,v的循环中生成一个新的字典，简化了循环的写法
str1='i love china for ever!'
{item:itemno for itemno,item in enumerate(str1)} #这里注意，字典的key是item即字母，value是索引即字符串的下标，由于字母有重复，所以value值会覆盖
```

```python
l2 = ["hello",'we','know']
for i,element in enumerate(l2):
    print(i,element)
```

```python
#生成器函数，最重要的是yield关键字，每次执行到yield这里会函数会返回,直到下次迭代，会从yield这个位置和状态继续执行。
def g2_func(n):
    for i in range(n):
        yield i**2
        
g2=g2_func(10)
```

```python
type(g2)
```

```python
for item in g2:
    print(item)
```

### 什么是高阶函数？把函数作为参数传入，这样的函数称为高阶函数
- python中内置高阶函数
    + map 映射函数，对可迭代对象中每个元素执行参数中的函数
    + reduce 缩减函数，对可迭代对象中元素，成对的去执行函数操作，直到最终只剩下一个值
    + filter 过滤函数，对刻碟对对象中元素迭代，根据函数去判断，过滤掉不满足函数条件的元素
    + sorted
- 其中map/reduce/filter都是参数形式一样，第一个参数是函数对象，第二个参数是可迭代对象
 - 下面是例子

```python
l1 = [1,2,3,4,5]
list(map(lambda x:x**2,l1))
```

```python
l2 = [1,3,5,7,9]
list(map(lambda x,y:x+y,l1,l2))
```

```python
from functools import reduce #奇怪的设计，为什么 map，filter不需要导入functools包

num = reduce(lambda x,y:x*y ,l1)
print(num)
```

```python
#函数使用三元表达式
l3 = filter(lambda x:True if x%2 ==0 else False,l1)
print(list(l3))
```

### 嵌套函数 闭包 LEGB名字访问规则  装饰器

```python
#函数outter 内定义另一个函数inner1，称为函数嵌套
def outter():
    name='jack'#定义了，被下层函数使用
    age=24#定义了，被下层函数使用
    address='BJ'#定义了，被下层函数做返回值
    hobbie='fishing'#定义了但未被本函数和下层函数使用
    
    def inner1():
        new_name=name[0:3]
        new_age+=age
        inner1_l=locals()
        inner1_g=globals()
        return address,inner1_l#使用outer的
    return inner1
inner1 = outter()
```

- 首先给出闭包函数的必要条件：
 - 闭包函数必须返回一个函数对象
 - 闭包函数返回的那个函数必须引用外部变量（一般不能是全局变量），而返回的那个函数内部不一定要return

- LEGB  locals --> encloure --> globals--> buildins

- 装饰器， 是一个python 语法糖。
- 在定义函数fun时，在其上一行写@outer_funcName。这样每次在调用fun()时，会把fun函数做为参数传递给outer_funcName函数，并执行该函数。

```python
import datetime

def outer(func):
    def inner():
        print('executed:',func())
        print(datetime.datetime.now())
    return inner

@outer
def fun1():
    return 'this is fun1 function--'


fun1()
```
